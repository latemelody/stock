```python
from datetime import datetime, timedelta, timezone
import pandas as pd
import matplotlib.pyplot as plt
from fbprophet import Prophet
import schedule
import time
import csv
```

```python
#data_a = pd.read_csv('stock_data/data/4661-オリエンタルランド.csv', usecols=[1,2,3,4,5,6])
data = pd.read_csv('stock_data/data/4661-オリエンタルランド.csv', usecols=[1, 5])
data.columns = ['ds', 'y']
```

```python
model = Prophet(daily_seasonality=True)
#Prophetのモデルを作成(daily_seasonality=Trueで1日単位の値動きの傾向も見る)

# 非線形トレンドの指定はこっち
#model = Prophet(daily_seasonality=True, growth='logistic')

model.fit(data)
#モデルにデータを流し込む
```

```python
future_data = model.make_future_dataframe(periods=365, freq = 'd')
#model.make_future_dataframe()で予測する期間を指定,periods=で予測する日数
forecast_data = model.predict(future_data)
#今後1年の株価の値動きを予測する関数を定義
```

```python
fig = model.plot(forecast_data)
model.plot_components(forecast_data)
#予測結果をプロット
```
